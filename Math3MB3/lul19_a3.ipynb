A cannon ball is shot from a cannon with an exit velocity 120 at angle 60 above parallel. 

Let $x$ be the horizontal position of the ball, and $y$ be the vertical position of the ball. Assume the ball is shot from a start position of $x=0, y=0$ at $t=0$. The air resistence is -0.02 in both directions. Wind is negligable for now. 

Use Euler's method with a step size of $h=1e-3$ and $n = 40000$. Remember that the vertical displacement cannot be less than 0 at any point in time. 

You need to account for gravity as well. 


`question_1` takes in NO arguements, calculates the positions of x and y over 40,000 points using Euler's method as previously covered in class. 

Once completed it returns a `list` with two entries of `np.array` type.
Each of those `np.array` entries must have a shape of `(40000,)`. 

The first entry should correspond to the $x$ values over time. 

The second entry of the list should correspond to the $y$ values over time. 

I have created visible tests for you to follow. 



```python
def question_1(): 

    n=40000
    x_hist = np.zeros(n)
    y_hist = np.zeros(n)
    vx_hist = np.zeros(n)
    vy_hist = np.zeros(n)

    gamma_x=-0.02
    gamma_y=-0.02
    eta_wind=0
    grav_const=-9.81
    ax = gamma_x+eta_wind
    ay = gamma_y+grav_const

    x_o=0
    y_o=0
    x_hist[0]=x_o
    y_hist[0]=y_o

    pow_x=60
    pow_y=60*np.sqrt(3)
    h=1e-3
    vx_hist[0]=pow_x
    vy_hist[0]=pow_y
    for i in range (0,n-1):
        x_hist[i+1] = x_hist[i]+vx_hist[i]*h
        y_hist[i+1] = y_hist[i]+vy_hist[i]*h
        if y_hist[i+1] <= 0:
            y_hist[i+1] = 0
        vx_hist[i+1] = vx_hist[i]+ax*h
        vy_hist[i+1] = vy_hist[i]+ay*h
        answer=[x_hist,y_hist]
    return(answer)

```

### LORENZ ATTRACTOR ( 6 points)

Using the code covered in class on the lorenz attractor, calculate the number of seconds spent in the cube bounded by 
$$-10 < x < -5, $$

$$-10 < y < -5,$$

$$25 < z < 30 $$. 

Remember that $h$ is the step-size with respect to time, so for every iteration in euler's method, you are stepping through time with $h$. 

```python
def question_lorenz():
    
    # I give you some starting values and code, what a nice instructor. 
    
    s=10
    r=28
    b=8/3
    n = 100000
    x_history = np.zeros(n)
    y_history = np.zeros(n)
    z_history = np.zeros(n)


    ic_x = 0
    ic_y = 1 
    ic_z = 1.05

    h = 1e-3

    x_history[0] = ic_x
    y_history[0] = ic_y
    z_history[0] = ic_z
    
    
    seconds_spent_in_cube = 0
    
    # YOUR CODE HERE
    from mpl_toolkits.mplot3d import Axes3D
    def diff_x(x,y,s_in = s):
        return(s_in*(y-x))
    def diff_y(x,y,z,r_in=r):
        return(r_in*x - y - x*z)
    def diff_z(x,y,z,b_in = b):
        return(x*y-b_in*z)

    for i in range (0,n-1):
        x_i=x_history[i]
        y_i=y_history[i]
        z_i=z_history[i]
        x_history[i+1]=x_history[i]+diff_x(x_i,y_i)*h
        y_history[i+1]=y_history[i]+diff_y(x_i,y_i,z_i)*h
        z_history[i+1]=z_history[i]+diff_z(x_i,y_i,z_i,b_in=b)*h

        if -10<x_history[i+1]<-5 and -10<y_history[i+1]<-5 and 25<z_history[i+1]<30:
            seconds_spent_in_cube +=h
            

    
    
    
    ax = pyplot.axes(projection='3d')
    ax.view_init(30, -35)
    ax.plot3D(x_history,y_history,z_history,linewidth = 0.5)
    
    return(seconds_spent_in_cube)

question_lorenz()
```
